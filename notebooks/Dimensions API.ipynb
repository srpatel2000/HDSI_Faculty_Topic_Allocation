```python
faculty = pd.read_csv('hdsi_faculty.csv')
display(faculty.head(3))

# grab the research ids
faculty_ids = faculty[faculty['Dimensions ID'] != 'no ID']['Dimensions ID'].to_list()
print(faculty_ids)
print(len(faculty_ids))
```

```python
dsl = dimcli.Dsl() # instantiation dsl
query_results = []
for i in faculty_ids:
    i = json.dumps(i)
    publication = dsl.query_iterative(f"""search publications where researchers.id = {i} and year > 2015 return publications[category_for+concepts+authors+date+year+id+journal+times_cited+title+abstract]""")
    query_results.append(publication.as_dataframe())
```

```python
pd.concat(query_results)
```

```python
pd.concat(query_results).to_csv("dimensions_siddhi.csv")
```
